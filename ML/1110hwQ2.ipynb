(a) Based on this data set, provide an estimate for the population mean of <font color="brown">medv</font>. Call this estimate $\hat{µ}$.

```python
print(f"estimate of population mean of medv: {Boston["medv"].mean():.4f}")
```

(b) Provide an estimate of the standard error of $\hat{µ}$. Interpret this result.<br>
_Hint: We can compute the standard error of the sample mean by dividing the sample standard deviation by the square root of the number of observations._

```python
import numpy as np
y = Boston["medv"].values

# Compute standard deviation
std = np.std(y, ddof=1)

# Number of observations
n = len(y)

# Standard error of sample mean
se_mean = std / np.sqrt(n)

print(f"Sample standard deviation: {std:.4f}")
print(f"Standard error of mean: {se_mean:.4f}")
```

(c) Now estimate the standard error of $\hat{µ}$ using the bootstrap. How does this compare to your answer from (b)?

```python
s = np.std(y, ddof=1)
se_analytic = s / np.sqrt(n)

# Bootstrap setting
B = 10000
rng = np.random.default_rng(314657018)  
boot_means = np.empty(B)

for b in range(B):
    sample = rng.choice(y, size=n, replace=True)
    boot_means[b] = sample.mean()

se_boot = np.std(boot_means, ddof=1)

print(f"n = {n}")
print(f"Sample SD (s): {s:.4f}")
print(f"Analytic SE = s / sqrt(n): {se_analytic:.4f}")
print(f"Bootstrap SE (B={B}): {se_boot:.4f}")
print(f"Absolute difference: {abs(se_boot - se_analytic):.6f}")
print(f"Relative difference: {abs(se_boot - se_analytic)/se_analytic:.4%}")
```

(d) Based on your bootstrap estimate from (c), provide a 95 % confidence interval for the mean of <font color="brown">medv</font>. Compare it to the results obtained by using <font color="brown">Boston['medv'].std()</font> and the two standard error rule (<font color="blue">3.9</font>).

_Hint: You can approximate a 95% confidence interval using the formula $[\hat{µ}− 2SE(\hat{µ}),\hat{µ} + 2SE(\hat{µ})]$._

```python
mu_hat = y.mean()

# 95% CI approximate using +/- 2 * SE_boot
ci_lower = mu_hat - 2 * se_boot
ci_upper = mu_hat + 2 * se_boot

print(f"95% CI by bootstrap (mu_hat ± 2 * SE_boot): [{ci_lower:.4f}, {ci_upper:.4f}]")
```

(e) Based on this data set, provide an estimate, $\hat{µ}_{med}$, for the median value of <font color="brown">medv</font> in the population.

```python
print(f"estimate of population median of medv: {Boston["medv"].median():.4f}")
```

(f) We now would like to estimate the standard error of $\hat{µ}_{med}$. Unfortunately, there is no simple formula for computing the standard
error of the median. Instead, estimate the standard error of the median using the bootstrap. Comment on your findings.

```python
# bootstrap
B = 10000
rng = np.random.default_rng(314657018)
boot_medians = np.empty(B)
for b in range(B):
    sample = rng.choice(y, size=n, replace=True)
    boot_medians[b] = np.median(sample)

se_median_boot = np.std(boot_medians, ddof=1)

# optional: bootstrap percentile 95% CI for the median
lower_pct, upper_pct = np.percentile(boot_medians, [2.5, 97.5])

print(f"n = {n}")
print(f"Bootstrap SE of median (B={B}): {se_median_boot:.4f}")
print(f"Bootstrap 95% percentile CI for median: [{lower_pct:.4f}, {upper_pct:.4f}]")
```

(h) Use the bootstrap to estimate the standard error of $\hat{µ}_{0.1}$. Comment on your findings.

```python
B = 10000
rng = np.random.default_rng(314657018)

boot_p10 = np.empty(B)
n = len(y)

for b in range(B):
    sample = rng.choice(y, size=n, replace=True)
    boot_p10[b] = np.percentile(sample, 10)

se_p10 = np.std(boot_p10, ddof=1)

print(f"Bootstrap SE of 10th percentile estimate: {se_p10:.4f}")
```
