(b) Create a diagram similar to the left-hand panel of Figure 8.14, using the tree illustrated in the right-hand panel of the same figure. You should divide up the predictor space into the correct regions, and indicate the mean for each region.

```python
import matplotlib.pyplot as plt
import numpy as np

with plt.xkcd():
    # Basic setup
    plt.rcParams['font.family'] = 'DejaVu Sans'
    fig, ax = plt.subplots(figsize=(8,6))

    rng = np.random.default_rng(12)
    x1 = rng.uniform(0, 2, 200)
    x2 = rng.uniform(0, 2, 200)

    ax.set_xlabel('$X_1$')
    ax.set_ylabel('$X_2$')

    # Cut points
    t1, t2, t3, t4 = 1, 1, 2, 0

    # Axis config
    xlim = [-1, 3]
    ylim = [-1, 3]
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    ax.set_xticks([0, 1])
    ax.set_yticks([1, 2])

    # For normalization (line endpoints)
    xrange = xlim[1] - xlim[0]
    yrange = ylim[1] - ylim[0]

    # Horizontal & vertical lines
    ax.axhline(t1, c='b', label='t1')
    ax.axhline(t3, c='b', label='t3')
    ax.axvline(t2, color='r',
               ymax=(t1-ylim[0]) / yrange,
               label='t2')
    ax.axvline(t4, color='r',
               ymin=(t1-ylim[0]) / yrange,
               ymax=(t3-ylim[0]) / yrange,
               label='t4')

    # --- Annotate line labels ---
    ax.text(t2, ylim[0] - 0.15, f'$t_2={t2}$',
            ha='center', va='center', color='k')
    ax.text(t4, t1 - 0.15, f'$t_4={t4}$',
            ha='center', va='center', color='k')

    ax.text(xlim[1] + 0.05, t1, f'$t_1={t1}$',
            rotation=90, ha='left', va='center', color='k')
    ax.text(xlim[1] + 0.05, t3, f'$t_3={t3}$',
            rotation=90, ha='left', va='center', color='k')

    # --- Annotate regions (use absolute coordinates instead of Axes fraction) ---
    region_labels = {
        (0.2, 0.3): '-1.80',
        (1.3, 0.3): '0.63',
        (0.8, 2.3): '2.49',
        (-0.5, 1.4): '-1.06',
        (1.0, 1.4): '0.21'
    }

    for (x, y), text in region_labels.items():
        ax.text(x, y, text, c='purple', fontsize=12)

    plt.title("Piecewise Constant Regions with Cut Points")
    plt.show()
```
