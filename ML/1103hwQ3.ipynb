(f) Create a function, PlotPower(), that allows you to create a plot of x against x**a for a fixed a and a sequence of values of x. For instance, if you call

PlotPower (np.arange (1, 11) , 3)

then a plot should be created with an x-axis taking on values 1, 2, . . . , 10, and a y-axis taking on values 1^3, 2^3, ...... , 10^3.

```python
import matplotlib.pyplot as plt
import numpy as np

def PlotPower(x, a, log=''):
    y = x ** a  # <-- define y before plotting

    fig, ax = plt.subplots(figsize=(8, 6))
    ax.plot(x, y, marker='o')
    ax.set_xlabel('x')
    ax.set_ylabel(f'x^{a}')
    ax.set_title(f'x^{a} vs x')

    # handle log scaling options
    if log == 'xy':
        ax.set_xscale('log')
        ax.set_yscale('log')
        ax.set_title(f'x^{a} vs x (both axes log scale)')
        ax.set_xlabel('log(x)')
        ax.set_ylabel(f'log(x^{a})')

    elif log == 'x':
        ax.set_xscale('log')
        ax.set_title(f'x^{a} vs log(x)')
        ax.set_xlabel('log(x)')

    elif log == 'y':
        ax.set_yscale('log')
        ax.set_title(f'log(x^{a}) vs x')
        ax.set_ylabel(f'log(x^{a})')

    plt.show()
```

```python
PlotPower(np.arange (1, 11) , 3)
```
