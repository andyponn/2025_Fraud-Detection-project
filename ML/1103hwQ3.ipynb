```python
import matplotlib.pyplot as plt

# 定義 Power3 函數
def Power3(x, a):
    result = x**a
    return result

# 建立 x 值（1 到 10 的整數）
x_values = range(1, 11)

# 計算 y = x^2
y_values = [Power3(x, 2) for x in x_values]

# 建立圖形與座標軸
fig, ax = plt.subplots(figsize=(6,4))

# 畫線圖
ax.plot(x_values, y_values, marker='o', linestyle='-', color='b', label=r'$f(x)=x^2$')

# 設定標題與軸標籤
ax.set_title('Plot of $f(x) = x^2$')
ax.set_xlabel('x')
ax.set_ylabel('y = x^2')

# 若要使用對數刻度，可以取消註解以下其中一行或兩行：
# ax.set_xscale('log')   # x 軸對數尺度
# ax.set_yscale('log')   # y 軸對數尺度

# 顯示圖例
ax.legend()

# 顯示圖形
plt.show()
```

(f) Create a function, PlotPower(), that allows you to create a plot of x against x**a for a fixed a and a sequence of values of x. For instance, if you call

PlotPower (np.arange (1, 11) , 3)

then a plot should be created with an x-axis taking on values 1, 2, . . . , 10, and a y-axis taking on values 1^3, 2^3, ...... , 10^3.

```python
import numpy as np
import matplotlib.pyplot as plt

# 原先的 Power3 函數
def Power3(x, a):
    return x**a

# 定義 PlotPower 函數
def PlotPower(x_values, a):
    """
    Plot x against x**a for a sequence of x values and a fixed exponent a.
    
    Parameters:
    x_values: array-like, sequence of x values
    a: float or int, exponent
    """
    # 計算 y 值
    y_values = [Power3(x, a) for x in x_values]
    
    # 建立圖形
    fig, ax = plt.subplots(figsize=(6,4))
    
    # 畫圖
    ax.plot(x_values, y_values, marker='o', linestyle='-', color='b', label=fr'$x^{a}$')
    
    # 標題與軸標籤
    ax.set_title(fr'Plot of $x^{a}$')
    ax.set_xlabel('x')
    ax.set_ylabel(fr'y = x^{a}')
    
    # 可選：對數尺度（若需要可取消註解）
    # ax.set_xscale('log')
    # ax.set_yscale('log')
    
    # 顯示圖例
    ax.legend()
    
    # 顯示圖形
    plt.show()

# 範例呼叫
PlotPower(np.arange(1, 11), 3)
```
